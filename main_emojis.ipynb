```python
# -*- coding: utf-8 -*-
import sqlite3
import pandas as pd
import csv
%matplotlib inline
import matplotlib.pyplot as plt
from translate import Translator
```

```python
conn = sqlite3.connect("Urban Data Hackathon/UTSEUS-shanghai-weibo-2012.db")
c = conn.cursor()
```

```python
req = "SELECT ws.weibo_location_id, status\
      FROM weibo_status AS ws, weibo_location AS wl\
      WHERE wl.weibo_location_id = ws.weibo_location_id"

c.execute(req)
raw_data = c.fetchall()
```

```python
labels = ['loc_id','status']

df = pd.DataFrame.from_records(raw_data,columns=labels)
df.shape[0]
```

```python
emoji_status = df[df.status.str.contains("\[")]
emoji_status.shape[0]
```

```python
emoji_status.to_csv('emoji_status.csv')
```

```python
from operator import itemgetter
filename = 'emoji_status.csv'
csv_file = csv.reader(open(filename, 'r',encoding='UTF-8'))
dict_cn = {}
for row in csv_file:
    #print(row)
    count = 0
    emoji = ''
    flag = False
    for ch in row[2]:
        if count >= 2:
            break
        if ch == ']':
            flag = False
            if emoji in dict_cn.keys():
                dict_cn[emoji] += 1
            else:
                dict_cn[emoji] = 1
            emoji=""
            count+=1
        else:
            if not flag:
                if ch == '[':
                    flag = True
            else:
                emoji += ch


                
dict_cn = {emoji:v for emoji, v in dict_cn.items() if ((v >50) & (emoji != ''))}                
dict_cn= sorted(dict_cn.items(), key=lambda d:d[1], reverse = True)
print(dict_cn)
```

```python
df = pd.DataFrame.from_dict(dict_cn)
df
```
